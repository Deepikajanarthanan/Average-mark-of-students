```python
import pandas as pd
```

```python
data=pd.read_csv(r"/content/catmarks.csv")
```

```python
def best_two_avg(row):
    marks = [row['Cat1'], row['Cat2'], row['Cat3']]
    marks.sort(reverse=True)
    best_two_sum = marks[0] + marks[1]
    average = best_two_sum / 8
    return average
data["Average"] = data.apply(best_two_avg, axis=1)
print(data)
data["Result"] = data["Average"].apply(lambda x: "Pass" if x >= 5 else "Fail")

result_counts = data["Result"].value_counts()
print("Number of students Pass/Fail:\n", result_counts)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,6))
plt.pie(result_counts, labels=result_counts.index, autopct='%1.1f%%', colors=["green","red"], startangle=90)
plt.title("Students Pass vs Fail (Based on Average Marks)")
plt.show()
```
